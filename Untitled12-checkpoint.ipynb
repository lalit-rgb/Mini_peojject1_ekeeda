### ML Model

```python
X = df_scaled.loc[:,['age','sex','bmi','children','smoker','region','charges']]
Y = df_scaled.loc[:,['insuranceclaim']]
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
dt_classifier = DecisionTreeClassifier()
dt_classifier.fit(X_train, Y_train)
```

```python
plt.figure(figsize=(15,10))
tree.plot_tree(dt_classifier, filled=True)
```

```python
y_pred = dt_classifier.predict(X_test)
y_pred
```

```python
accuracy = accuracy_score(Y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
y_predt = dt_classifier.predict(X_train)
y_predt
```

```python
accuracy = accuracy_score(Y_train, y_predt)
print("Accuracy:", accuracy)
```

```python
cm7 = confusion_matrix(Y_test, y_pred)
print(cm7)
fig, ax = plot_confusion_matrix(conf_mat=cm7, figsize=(6, 6), cmap=plt.cm.Greens)
plt.xlabel('Predictions', fontsize=18)
plt.ylabel('Actuals', fontsize=18)
plt.title('Confusion Matrix', fontsize=18)
plt.show()
```

```python
X = df_scaled.loc[:,['age','sex','bmi','children','smoker','region','charges']]
Y = df_scaled.loc[:,['insuranceclaim']]
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
dt_classifier = DecisionTreeClassifier(max_depth = 8)
dt_classifier.fit(X_train, Y_train)
```

```python
plt.figure(figsize=(15,10))
tree.plot_tree(dt_classifier, filled=True)
```

```python
y_pred = dt_classifier.predict(X_test)
y_pred
```

```python
accuracy = accuracy_score(Y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
y_predtt = dt_classifier.predict(X_train)
y_predtt
```

```python
accuracy = accuracy_score(Y_train, y_predtt)
print("Accuracy:", accuracy)
```

```python
cm7 = confusion_matrix(Y_test, y_pred)
print(cm7)
fig, ax = plot_confusion_matrix(conf_mat=cm7, figsize=(6, 6), cmap=plt.cm.Greens)
plt.xlabel('Predictions', fontsize=18)
plt.ylabel('Actuals', fontsize=18)
plt.title('Confusion Matrix', fontsize=18)
plt.show()
```

```python
print("Report : ",classification_report(Y_test, y_pred))
```

As we can see that in this classification problem Decision tree Machine Learning is performing really well here to check the purity by default this model is using GINI INDEX.

In our first attempt we can see that even though the accuracy is 97% it is overfitted as the training dataset as as prediction of 100% which shows that our data is completely overfitting. 

As we know that our data set has only 1338 data records which is considerably a small data set so we can use post prunning here.

So we have done the post Prunning on the data with max depth = 8 by doing that we have reduced the unnecessary splitting of the data. we could have taken the max depth = 7  but at the depth of 7 we have a node with gini index = 0.47(  gini index = 0.5 if the node is completely impure) so I had to go with a depth of 8. 

By doing this we have not reduced ur accuracy drastically but it went from 97% to 96.6% but now we have gotten rid of the overfitting as the accuracy on our training data is not 100% now.
